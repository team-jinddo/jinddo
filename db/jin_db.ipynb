## 테이블 생성(비식별키)

```python
def create_tables():
    connection = get_connection()
    try:
        with connection.cursor() as cursor:
            # user_tb: 사용자 정보
            user_tb = """
            CREATE TABLE IF NOT EXISTS user_tb (
                user_id INT AUTO_INCREMENT PRIMARY KEY,
                email VARCHAR(255) NOT NULL,
                password VARCHAR(255) NOT NULL,
                created_at DATETIME DEFAULT CURRENT_TIMESTAMP
            );
            """
            cursor.execute(user_tb)

            # rest_tb: 식당 정보
            rest_tb = """
            CREATE TABLE IF NOT EXISTS rest_tb (
                rest_id INT AUTO_INCREMENT PRIMARY KEY,
                name VARCHAR(255) NOT NULL,
                address VARCHAR(255) NULL,
                url VARCHAR(255) NULL
            );
            """
            cursor.execute(rest_tb)

            # review_tb: 방문자 리뷰
            review_tb = """
            CREATE TABLE IF NOT EXISTS review_tb (
                review_id INT AUTO_INCREMENT PRIMARY KEY,
                user_id INT,
                rest_id INT,
                author VARCHAR(20) NULL,
                review_content TEXT NULL,
                publication_date DATETIME NULL,
                CONSTRAINT fk_review_user
                    FOREIGN KEY (user_id) REFERENCES user_tb(user_id),
                CONSTRAINT fk_review_rest
                    FOREIGN KEY (rest_id) REFERENCES rest_tb(rest_id)
            );
            """
            cursor.execute(review_tb)

            # recommend_tb: 추천 기록
            recommend_tb = """
            CREATE TABLE IF NOT EXISTS recommend_tb (
                recommend_id INT AUTO_INCREMENT PRIMARY KEY,
                user_id INT NOT NULL,
                rest_id INT NOT NULL,
                score INT NULL,
                CONSTRAINT fk_recommend_user
                    FOREIGN KEY (user_id) REFERENCES user_tb(user_id),
                CONSTRAINT fk_recommend_rest
                    FOREIGN KEY (rest_id) REFERENCES rest_tb(rest_id)
            );
            """
            cursor.execute(recommend_tb)

            # taste_tb: 입맛 정보
            taste_tb = """
            CREATE TABLE IF NOT EXISTS taste_tb (
                taste_id INT AUTO_INCREMENT PRIMARY KEY,
                taste_type VARCHAR(255) NOT NULL
            );
            """
            cursor.execute(taste_tb)

            # category_tb: 음식 타입
            category_tb = """
            CREATE TABLE IF NOT EXISTS category_tb (
                category_id INT AUTO_INCREMENT PRIMARY KEY,
                category_name VARCHAR(255) NOT NULL
            );
            """
            cursor.execute(category_tb)

            # category_rec_tb: 음식타입 기록 중간 테이블 (사용자와 음식 타입 관계)
            category_rec_tb = """
            CREATE TABLE IF NOT EXISTS category_rec_tb (
                user_id INT,
                category_id INT,
                CONSTRAINT fk_category_rec_user
                    FOREIGN KEY (user_id) REFERENCES user_tb(user_id),
                CONSTRAINT fk_category_rec_category
                    FOREIGN KEY (category_id) REFERENCES category_tb(category_id)
            );
            """
            cursor.execute(category_rec_tb)

            # taste_rec_tb: 입맛 기록 중간 테이블 (사용자와 식당 타입 관계)
            taste_rec_tb = """
            CREATE TABLE IF NOT EXISTS taste_rec_tb (
                user_id INT,
                taste_id INT,
                CONSTRAINT fk_taste_rec_user
                    FOREIGN KEY (user_id) REFERENCES user_tb(user_id),
                CONSTRAINT fk_taste_rec_taste
                    FOREIGN KEY (taste_id) REFERENCES taste_tb(taste_id)
            );
            """
            cursor.execute(taste_rec_tb)

        connection.commit()  # 변경 사항 커밋
        print("All tables created successfully!")
    
    except Exception as e:
        print(f"An error occurred: {e}")
        connection.rollback()  # 에러 시 롤백
    
    finally:
        connection.close()

# 테이블 생성 함수 실행
create_tables()

```

## 식당 삽입 코드

```python
def insert_rest_data(df):
  # 데이터베이스 연결 설정
  connection = get_connection()
  try:
    with connection.cursor() as cursor:
      #여러 값을 삽입하기 위한 SQL 쿼리 준비
      sql = """
      INSERT INTO rest_tb (name, address, url)
      VALUES (%s, %s, %s)
      """
      # 삽입할 데이터 준비
      data = []
      for _, row in df.iterrows():
        name = row['사업장명']
        address = row['도로명주소']
        url = row['url']
        
        #데이터 추가
        data.append((name, address, url))
      # 모든 데이터를 한 번에 삽입
      cursor.executemany(sql, data)
      # 변경사항 커밋
      connection.commit()
  except Exception as e:
    print(f'Error occurred: {e}')
    #에러 발생 시 롤백
    connection.rollback()
  finally:
    # 연결 종료
    connection.close

df = df[['사업장명', '도로명주소', 'url']]

# 데이터 삽입 함수 호출
insert_rest_data(df)
```

## 유저 정보 테이블 데이터 삽입

```python
import random
import string
from datetime import datetime
```

```python
def random_password(length=10):
  characters = string.ascii_letters + string.digits
  return ''.join(random.choice(characters) for _ in range(length))

def convert_date(date_str):
    try:
        return datetime.strptime(date_str, '%Y. %m. %d. %H:%M') if pd.notna(date_str) else None
    except ValueError:
        return None
  
def insert_user_data(user_df):
  # 데이터베이스 연결 설정
  connection = get_connection()
  try:
    with connection.cursor() as cursor:
      sql = """
      INSERT INTO user_tb (email, password, created_at)
      VALUES (%s, %s, %s)
      """

      # 삽입할 데이터 준비
      data = []
      for _, row in user_df.iterrows():
        author_id = row['author_id']
        # 이메일 생성
        email = f"{author_id}@naver.com"
        password = random_password()
        created_date = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

        data.append((email, password, created_date))
      # 모든 데이터를 한 번에 삽입
      cursor.executemany(sql, data)

      # 변경사항 커밋
      connection.commit()
  except Exception as e:
    print(f"Error occurred: {e}")
    #에러 발생 시 롤백
    connection.rollback()
  finally:
    # 연결 종료
    connection.close

# user_df에서 author_id 열만 선택
user_df = review_data[['author_id']]

# 데이터 삽입 함수 호출
insert_user_data(user_df)
```
